```python
import tensorflow_hub as hub 
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model=hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3 )
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_image = load_image('C:/Users/ASUS/Downloads/test.jpeg')
style_image = load_image('C:/Users/ASUS/Downloads/style.jpeg')
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylized_image))
plt.show()
```

```python
cv2.imwrite('C:/Users/ASUS/Downloads/generated_img.jpg', cv2.cvtColor(np.squeeze(stylized_image)*255,cv2.COLOR_BGR2RGB))
```
